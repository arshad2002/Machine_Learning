```python
import glob
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
files = glob.glob('Data/image/CIFAR-10-images/train/**/*')
files
```

```python
classes = os.listdir('Data/image/CIFAR-10-images/train')
class_count =[]
for c in classes:
    file = glob.glob(str('Data/image/CIFAR-10-images/train/'+c+'/*'))
    class_count.append(len(file))
```

```python
plt.figure(figsize=(17,7))
sns.barplot(x=classes, y=class_count)
```
