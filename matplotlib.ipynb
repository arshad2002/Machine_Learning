### Workflow

```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig("image/simple-plot.png")
```

### Making figure with numpy array

```python
x = np.linspace(0, 10, 100)
x
```

```python
# line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# bar graph using dictionary
prices = {
    "chicken":100,
    "cow": 200,
    "duck": 150
}
fig, ax = plt.subplots()
ax.bar(prices.keys(), prices.values())
ax.set(title="Prices",
      ylabel="price",
      xlabel="goods");
```
